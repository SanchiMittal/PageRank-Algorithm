```python
import graph
from page_rank import get_all_ranks
total_nodes = 10 # total no. of nodes
total_edges = 30 # total no. of edges
d = 0.85 # damping factor
total_iterations = 3

# Create an object of Graph 
g = graph.generate_graph(total_nodes, total_edges)
ranks = [1.0 / total_nodes] * total_nodes
print("inital ranks =  ", ranks)

final_ranks, rank_sums = get_all_ranks(ranks, g.in_link_nodes, g.out_degree, d, total_nodes, total_edges, total_iterations)

for i in range(total_iterations):
    print("\nIteration", i, ":")
    print("Page ranks = ", final_ranks[i])
    print("Rank Sum = ", rank_sums[i])
```
